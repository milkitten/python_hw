```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
# load data 
data_dir = "data.txt"
train = pd.read_csv(data_dir, sep = ' ', header = 0)
print(train.columns)
train = train.sort_values(by='context_timestamp', ascending=False)
train['context_timestamp']
```

```python
# 计算用户的历史转化率，广告的历史转化率（发生交易比例）
# 分別對 user_id 和 item_id 做 group by 求轉化率
group_item_id = train.groupby('item_id')
group_user_id = train.groupby('user_id')
print(group_item_id.size());
print(group_user_id.size());
item_cv = group_item_id['is_trade'].mean();
user_cv = group_user_id['is_trade'].mean();
```

```python
# 查看前五項結果
print(item_cv.head())
print(user_cv.head())

# filer非零轉化率的廣告/商品
idxs = np.nonzero([i > 0 for i in user_cv])
print(len(idxs[0]))
```

```python
# 计算每条记录中，当期用户距离上次浏览当前广告的时长
group = train.groupby(['user_id','item_id'])
diff_time = group['context_timestamp'].diff()
# diff_time[~np.isnan(diff_time)]
diff_time
```
